# Olympics Country Wise medals tally Bar Graph

#### Importing olympics medals csv file using pandas read csv function

```python
df = pd.read_csv("/kaggle/input/paris-2024-olympic-summer-games/medals.csv")
df
```

#### Checking if there is any null values in the dataframe

```python
df.isna().count()
```

##### Looks like it doesn't have any null values

#### Finding out the total number of medals each country got using value_counts function of pandas

```python
data1 = pd.value_counts(df['country'])
data2 = pd.DataFrame(data1)
data2
```

#### Converting the dataframe into a bar graph

```python
data1.plot.bar(x = 'country',y = 'count', rot = 90)
plt.rcParams['figure.figsize'] = (30,10)
# for i in range (len('country')):
#     plt.text(i,df.count[i],df.count[i])
```

### We can observe that United States has the highest medal tally followed by China and Great Britain
